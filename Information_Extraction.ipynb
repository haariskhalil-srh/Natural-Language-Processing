```python
import spacy
from spacy import displacy
from IPython.display import HTML, display

#   Load the English language model
nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

# Print the dependency parse information for each token
for token in doc:
    print(f"{token.text:<12} {token.dep_:<12} {token.head.text}")

html = displacy.render(doc, style="dep", jupyter=False)

# Display the dependency parse visualization in Jupyter Notebook
display(HTML(html))

# Discuss with your partner:
# 1. What is the ROOT?
# 2. What does "with a telescope" modify?
# 3. Does the sentence have more than one meaning?
# 4. Did spaCy choose the interpretation you expected?
# 5. Rewrite the sentence to remove ambiguity.
```

### 1. What is the ROOT?
The ROOT is the main verb of the sentence, which is "saw". Everything else structurally branches out from this word.

### 2. What does "with a telescope" modify?
The preposition "with" modifies the verb "saw". The model has classified it as the instrument used to perform the action of seeing.

### 3. Does the sentence have more than one meaning?
Yes. This is a famous example of syntactic ambiguity called prepositional phrase attachment ambiguity. It can mean two different things:
* **Meaning A:** I used a telescope to look at the man.
* **Meaning B:** I looked at a man, and that man was carrying a telescope.

### 4. Did spaCy choose the interpretation you expected?
Yes.

### 5. Rewrite the sentence to remove ambiguity.
* To force Meaning A: "Using a telescope, I saw the man."
* To force Meaning B: "I saw the man who was holding a telescope."

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative
in Berlin on 15 June 2026.

The project involves collaboration with Google,
SAP and SRH University Heidelberg.

Satya Nadella will present the initiative
at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(
        ent.text.ljust(30),
        ent.label_.ljust(15),
        spacy.explain(ent.label_)
    )
```

### 1. Which entities were detected?
The model detected the following entities:
* Microsoft
* Berlin
* 15 June 2026
* Google
* SAP
* AI
* Satya Nadella
* Germany

### 2. Which labels were assigned?
* **ORG:** Microsoft, Google, SAP, AI.
* **GPE:** Berlin, Germany
* **DATE:** 15 June 2026
* **PERSON:** Satya Nadella

### 3. Which entity types occur most often?
**ORG** is the most frequent label. It appears 4 times in the text.

### 4. Were any entities missed?
Yes. The model completely missed **SRH University Heidelberg**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))
```

### 1. Which pair has the highest similarity?
Doc1 and Doc2 have the highest similarity score.

### 2. Is this result reasonable?
Not resonable because the stopword "is" causes the similarity score of Doc1 Vs Doc3 to able be the same as Doc1 vs Doc2, even though they have completely different meaning.

### 3. Why does TF-IDF treat some words as more important?
The IDF component specifically penalizes words that appear across many documents (like "is" or "a") because they carry no unique information. It boosts the mathematical weight of rare words (like "healthcare") that help distinguish one document from another.

### 4. Does cosine understand meaning?
No. Cosine similarity has zero understanding of language.

### 5. Would it know that "AI" and "machine learning" are related?
No. Both words are completely different. A simple math calculation like cosine similarity cannot understand context.
